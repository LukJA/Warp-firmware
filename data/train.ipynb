```python
### Compute the probabilities of two tests where walking is class[1] == '1'
walking_test = walking.drop(['Class'], axis=1).tail(1)
stairs_test = stairs.drop(['Class'], axis=1).tail(1)
none_test = none.drop(['Class'], axis=1).tail(1)
```

```python
from scipy.special import logsumexp
from pprint import pprint
import math

# precomputions
class_0_prior,class_1_prior = gnb.class_prior_[0], gnb.class_prior_[1]
theta_0, theta_1 = gnb.theta_[0, :], gnb.theta_[1, :]
var_0, var_1 = gnb.var_[0, :], gnb.var_[1, :]
jointi_0 = np.log(class_0_prior)
jointi_1 = np.log(class_1_prior)
nij_0_p = -0.5*np.sum(np.log(2.0 * np.pi * var_0)) 
nij_1_p = -0.5*np.sum(np.log(2.0 * np.pi * var_1))

def test(X):
    # Class 0
    n_ij_0 = nij_0_p - 0.5*np.sum(((X - theta_0) ** 2) / (var_0), 1)

    # Class 1
    n_ij_1 = nij_1_p - 0.5 * np.sum(((X - theta_1) ** 2) / (var_1), 1)

    jll = np.array([jointi_0 + n_ij_0, jointi_1 + n_ij_1]).T

    # normalize by P(x) = P(f_1, ..., f_n)
    # log_prob_x = np.log(np.sum(np.exp(jll)))
    # p_walk = math.exp(jll[0][1] - log_prob_x)
    # print(f"P(Walking) = {p_walk:.3f}")

    p_walk = math.exp(jll[0][1]  - np.max(jll))

    if (nij_0_p == np.int0(nij_0_p)):
        print("Approximate Result")
    print(f"P(Walking) = {p_walk:.3f}")


test(walking_test)
test(stairs_test)
test(none_test)

```

## Approximated Math wihtout logarithms 

- For Feature Values $0 \rightarrow n$ = $X_n$ integers
- For Class 0 (other) Variance $V_{0n}$, Theta $\Theta_{0n}$, Baseline $N_0$
- For Class 0 (other) Variance $V_{1n}$, Theta $\Theta_{1n}$, Baseline $N_1$
- With a joint distribution set to 0 each

$ N_{0ij} = N_0 - \frac{1}{2} \sum_n (X_n - \Theta_{0n})^2 / V_{0n}$ 

$ N_{1ij} = N_1 - \frac{1}{2} \sum_n (X_n - \Theta_{1n})^2 / V_{1n}$

$ P_{1L} = N_{1ij} - MAX(N_{0ij}, N_{1ij})$

$ P_1 = exp\{ P_{1L} \}$






```python
def test_A(X):
    # Class 0
    n_ij_0 = np.int0(nij_0_p - 0.5*np.sum(((X - theta_0) ** 2) / (var_0), 1))

    # Class 1
    n_ij_1 = np.int0(nij_1_p - 0.5 * np.sum(((X - theta_1) ** 2) / (var_1), 1))

    p_walk_log = int(n_ij_1  - max(n_ij_1[0], n_ij_0[0]))
    p_walk = math.exp(p_walk_log)

    if (nij_0_p == np.int0(nij_0_p)):
        print("-- Approximate Result")
        
    print(f"P(Walking) = {p_walk:.2f}, P(log) = {p_walk_log:.2f}")

test_A(walking_test)
test_A(stairs_test)
test_A(none_test)
```

```python
def approx_exp(x, n = 12):
    res = 100
    # print("1 ?+ ")?
    factorial = 1
    for i in range(1, n):
        factorial *= i
        # print(f" X^{i} \ {factorial}")
        print(factorial)
        res += x**i * (100/factorial)
    if res < 0:
        return 0
    return res

for i in range(-10, 1):
    print(f"{approx_exp(i, 12):.3f}")
    print("")

```
